#### üáµüá≠ 1. Republic Act No. 8749 ‚Äì **Philippine Clean Air Act of 1999**  
üîó [RA 8749 Full Text](https://lawphil.net/statutes/repacts/ra1999/ra_8749_1999.html)

**Purpose:**  
To maintain clean air that meets national standards and to protect the public from the harmful effects of air pollutants.

**Key Provisions:**  
- Establishment of ambient air quality standards  
- Prohibition of incineration and open burning  
- Mandatory emission testing for vehicles and industries  
- Creation of airshed governing bodies

---

#### üíß 2. Republic Act No. 9275 ‚Äì **Philippine Clean Water Act of 2004**  
üîó [RA 9275 Full Text](https://lawphil.net/statutes/repacts/ra2004/ra_9275_2004.html)

**Purpose:**  
To protect the country‚Äôs water bodies from pollution through a comprehensive water quality management program.

**Key Provisions:**  
- Classification of water bodies (Class A, B, C, etc.)  
- Wastewater discharge permits for businesses  
- Establishment of Water Quality Management Areas (WQMAs)  
- Mandates DENR and LGUs to monitor and improve water quality

---

#### üóëÔ∏è 3. Republic Act No. 9003 ‚Äì **Ecological Solid Waste Management Act of 2000**  
üîó [RA 9003 Full Text](https://lawphil.net/statutes/repacts/ra2000/ra_9003_2000.html)

**Purpose:**  
To institutionalize the segregation, recycling, and composting of solid waste at the source.

**Key Provisions:**  
- Waste segregation at the household level  
- LGU-led Materials Recovery Facilities (MRFs)  
- Mandatory waste reduction targets  
- Prohibits open dumping of waste

---

#### ‚ò£Ô∏è 4. Republic Act No. 6969 ‚Äì **Toxic Substances and Hazardous and Nuclear Wastes Control Act of 1990**  
üîó [RA 6969 Full Text](https://lawphil.net/statutes/repacts/ra1990/ra_6969_1990.html)

**Purpose:**  
To regulate and control the importation, manufacture, use, and disposal of toxic substances and hazardous/nuclear waste.

**Key Provisions:**  
- Mandatory registration of chemicals  
- Proper labeling and handling of hazardous materials  
- Licensing for hazardous waste generators and transporters  
- Safe e-waste disposal policies

---

#### üìã 5. Presidential Decree No. 1586 ‚Äì **Environmental Impact Statement (EIS) System**  
üîó [PD 1586 Full Text](https://lawphil.net/statutes/presdecs/pd1978/pd_1586_1978.html)

**Purpose:**  
To require Environmental Impact Assessments (EIA) for projects that may significantly affect the environment.

**Key Provisions:**  
- Projects must secure an Environmental Compliance Certificate (ECC)  
- Establishment of procedural and documentation standards  
- Prior consultation with affected communities

---

#### üåø 6. Presidential Decree No. 1151 ‚Äì **Philippine Environmental Policy (1977)**  
üîó [PD 1151 Full Text](https://lawphil.net/statutes/presdecs/pd1977/pd_1151_1977.html)

**Purpose:**  
To declare environmental protection as a national policy and promote harmony between development and nature.

**Key Provisions:**  
- Duty of all citizens to contribute to environmental protection  
- Mandates government agencies to assess environmental impact  
- Recognizes inter-generational responsibility

---

#### üåä 7. Presidential Decree No. 979 ‚Äì **Marine Pollution Decree of 1976**  
üîó [PD 979 Full Text](https://lawphil.net/statutes/presdecs/pd1976/pd_979_1976.html)

**Purpose:**  
To prevent and control marine pollution from ships, offshore platforms, and coastal sources.

**Key Provisions:**  
- Prohibits dumping of pollutants into Philippine waters  
- Requires containment and cleanup systems for spills  
- Imposes penalties for maritime pollution

---

#### üíß 8. Presidential Decree No. 1067 ‚Äì **Water Code of the Philippines (1976)**  
üîó [PD 1067 Full Text](https://lawphil.net/statutes/presdecs/pd1976/pd_1067_1976.html)

**Purpose:**  
To regulate water use, rights, conservation, and infrastructure development.

**Key Provisions:**  
- Priority rights in water use (domestic > irrigation > power > others)  
- Permits required for water extraction and use  
- Protection of water sources from depletion and contamination

---

#### üöø 9. Presidential Decree No. 856 ‚Äì **Code on Sanitation of the Philippines (1975)**  
üîó [PD 856 Full Text](https://lawphil.net/statutes/presdecs/pd1975/pd_856_1975.html)

**Purpose:**  
To protect and promote public health through proper sanitation practices.

**Key Provisions:**  
- Regulations for food establishments, water quality, sewage, waste disposal  
- Mandates sanitary permits and inspections  
- Covers public health hazards in schools and workplaces

---

#### ‚öñÔ∏è 10. Writ of Kalikasan (Rule of Procedure for Environmental Cases)  
üîó [Writ of Kalikasan Info](https://en.wikipedia.org/wiki/Writ_of_Kalikasan)

**Purpose:**  
To provide a legal remedy for the protection of the constitutional right to a healthy environment.

**Key Features:**  
- Issued by the Supreme Court or Court of Appeals  
- Covers environmental damage of **magnitude**  
- Available to individuals, NGOs, or communities
