# **Course Title:** Environmental Engineering  
**Course Code:** CpE 333  
**Course Credits:** 3 Units (3 hours lecture per week)  
**Program:** Bachelor of Science in Computer Engineering (BSCpE)  
**Course Type:** Academic  
**Year Level:** 3rd or 4th Year  
**Prerequisite(s):** Basic Programming, Electronics, Introduction to IoT, Data Analytics  

---

## Course Description  
This course introduces students to the fundamental principles of environmental engineering with an emphasis on applications in computer engineering. Topics include environmental pollution, sustainability, environmental laws, and the development of smart systems for environmental monitoring using IoT, sensors, and machine learning. The course aims to foster environmental consciousness and empower students to design tech-based solutions for environmental problems.

---

## Course Outcomes  
At the end of the course, students should be able to:  
1. Explain the core concepts and significance of environmental engineering in the context of computer engineering.  
2. Identify types and sources of environmental pollution and evaluate mitigation technologies.  
3. Apply appropriate sensors and IoT architectures for environmental data collection.  
4. Analyze environmental data and create predictive models using machine learning techniques.  
5. Design and present a prototype or concept of a smart environmental system or device.  
6. Demonstrate awareness of environmental regulations and ethical responsibilities.

---

## 12-Week Topic Outline

| Week | Topic | Intended Learning Outcomes | Teaching-Learning Activities | Assessment Methods |
|------|-------|-----------------------------|------------------------------|---------------------|
| 1 | Introduction to Environmental Engineering | Discuss environmental issues and sustainability goals | Lecture, discussion | Quiz, recitation |
| 2 | Air and Water Pollution | Identify pollution types and environmental impacts | Case analysis, group activity | Worksheet, group report |
| 3 | Waste Management and E-Waste | Explain e-waste and sustainable disposal | Green computing video analysis | Reflection paper |
| 4 | Environmental Laws and Standards | Apply environmental laws (e.g., RA 8749, RA 9275, ISO 14000) | Role-play, debates | Written test |
| 5 | Sensors for Environmental Monitoring | Demonstrate use of sensors for environmental variables | Hands-on lab (DHT11, MQ sensors, water sensors) | Lab report |
| 6 | IoT for Environmental Applications | Design basic IoT system architecture | IoT prototyping, TinkerCAD/Fritzing simulation | Lab practical |
| 7 | Smart Energy and Green Buildings | Analyze smart systems for energy efficiency | Site analysis, simulation | Energy audit activity |
| 8 | Data Acquisition and Analysis | Perform preprocessing and data visualization | Python/MATLAB exercises | Code submission |
| 9 | Machine Learning for Environmental Prediction | Implement simple ML model for forecasting | Use Scikit-learn or TensorFlow | ML mini-project |
| 10 | Edge Computing and Cloud Integration | Integrate edge and cloud services | Setup Firebase/ThingSpeak dashboards | Project integration task |
| 11 | Capstone Concept Design | Propose an IoT-based environmental solution | Group consultation and design pitch | Proposal presentation |
| 12 | Project Presentation and Impact Assessment | Present environmental tech prototype or plan | Final presentation | Rubric-based evaluation |

---

## Course Requirements  
- Quizzes and Written Exams  
- Laboratory Activities and Reports  
- Mini-Projects (ML, IoT)  
- Final Group Project with Presentation  
- Attendance and Participation
